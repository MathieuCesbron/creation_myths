# Imports

```python
import json
import matplotlib.pyplot as plt
%matplotlib inline

from collections import defaultdict
```

# Plot occurence of numbers

```python
data_json_path = "./data.json"
numbers = {}

with open(data_json_path, "r") as file:
    data = json.loads(file.read())

    for creation_myth in data["creation_myths"]:
        # total = sum(creation_myth["stats"]["numbers"].values())
        for number, occurence in creation_myth["stats"]["numbers"].items():
            numbers[number] = numbers.get(number, 0) + occurence
            # numbers[number] = (numbers.get(number, 0) + occurence) / total
```

```python
# Occurences of numbers from 1 to 10

# Extracting values for keys 0 to 10
values = [numbers.get(str(key), 0) for key in range(1, 11)]

# Keys ranging from 0 to 10
keys = list(range(1, 11))
plt.bar(keys, values)

plt.xticks(range(11))
plt.xlim(0.5, 10.5)

plt.xlabel('Numbers')
plt.ylabel('Counts')
plt.title('Occurences of each numbers 1-10 in all creation myths')

plt.show()
```

# Plot probability to see each number

```python
data_json_path = "./data.json"
numbers = defaultdict(int)
total = 0

with open(data_json_path, "r") as file:
    data = json.loads(file.read())

    for creation_myth in data["creation_myths"]:
        total += 1
        for number in creation_myth["stats"]["numbers"].keys():
            numbers[number] += 1

for key in numbers:
    numbers[key] /= total
```

```python
# Occurences of numbers from 1 to 10

# Extracting values for keys 0 to 10
values = [numbers.get(str(key), 0) for key in range(1, 11)]

# Keys ranging from 0 to 10
keys = list(range(1, 11))
plt.bar(keys, values)

plt.xticks(range(11))
plt.xlim(0.5, 10.5)

# Create custom y-axis tick labels as percentages
y_ticks = [f'{int(y * 100)}%' for y in plt.yticks()[0]]
plt.yticks(plt.yticks()[0], y_ticks)

plt.xlabel('Numbers')
plt.ylabel('Probabilty')
plt.title('Probability to see each number 1-10 in a creation myth')

plt.show()
```
